```python
#----------Step 1, Step 2 and Step 3 --------------
# For this question, the source URL wasn't working so downloaded and then used it

import pandas as pnd
import datetime as dat

data_receiver = '/Users/gourav/Downloads/file.csv'
data = pd.read_csv(data_receiver, sep = "\s+", parse_dates = [[0,1,2]]) 
data_receiver
```

```python
#----------Step 4 --------------


def fixing_year(input):
  year_con = input.year - 100 if input.year > 1989 else input.year
  return dat.date(year_con, input.month, input.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fixing_year)

data.head()
```

```python
#----------Step 5 --------------


data["Yr_Mo_Dy"] = pnd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy')
data.head()
```

```python
import pandas as pnd


import numpy as nmp
import matplotlib.pyplot as matplt
%matplotlib inline

us_marriageDivorce_data = pnd.read_csv('/Users/gourav/Downloads/us-marriages-divorces-1867-2014.csv')
us_marriageDivorce_data = us_marriageDivorce_data[
    us_marriageDivorce_data['Year'].apply(lambda x: x in [1900, 1950, 2000])]

years_data = us_marriageDivorce_data['Year'].values
marriagesPerCapita = us_marriageDivorce_data['Marriages_per_1000'].values
divorcesPerCapita = us_marriageDivorce_data['Divorces_per_1000'].values


matplt.figure()
matplt.bar(years_data, 
        marriagesPerCapita, 
        color='red',
        width=2.3,
        align='center',
        label='MarriagesPerCapita')

matplt.bar(years_data, divorcesPerCapita,
        color='green',
        width=4.9,
        align='center',
        label='DivorcesPerCapita')
matplt.xlabel("Years(1900 - 2000)")
matplt.ylabel("Number of Marriage-Divorce")

matplt.legend()
```
